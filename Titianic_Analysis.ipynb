```python
from sklearn.linear_model import LogisticRegression
import time
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import datetime
from sklearn.naive_bayes import GaussianNB
```

```python
def fit_algos(algo,X_train,y_train,cv):
    
    model=algo.fit(X_train,y_train)
    accuracy=round(model.score(X_train,y_train)*100)
    
    #Cross Validation
    
    train_pred=cross_val_predict(algo,
                                                 X_train,
                                                 y_train,
                                                 cv=cv,
                                                 n_jobs=-1
    )
    
    #Accuracy Score - cross validation
    
    acc_cv=round(accuracy_score(y_train,train_pred)*100,2)
    
    return train_pred,accuracy,acc_cv
```

```python
start_time=time.time()

train_pred_log,acc_log,acc_cv_log=fit_algos(LogisticRegression(),
                                            X_train,
                                            y_train,
                                            10
)

log_time=(time.time()-start_time)

print('Accuracy is ',acc_log)
print('Accuracy CV 10 Fold is ',acc_cv_log)
print('Ruuning time is ',datetime.timedelta(seconds=log_time))
```

```python
start_time=time.time()

train_pred_log,acc_log,acc_cv_log=fit_algos(KNeighborsClassifier(),
                                            X_train,
                                            y_train,
                                            10
)

knn_time=(time.time()-start_time)

print('Accuracy is ',acc_log)
print('Accuracy CV 10 Fold is ',acc_cv_log)
print('Ruuning time is ',datetime.timedelta(seconds=knn_time))
```

```python
import numpy as np
from catboost import CatBoostClassifier, Pool,cv
```

```python
start_time=time.time()

train_pred_log,acc_log,acc_cv_log=fit_algos(GaussianNB(),
                                            X_train,
                                            y_train,
                                            10
)

GB_time=(time.time()-start_time)

print('Accuracy is ',acc_log)
print('Accuracy CV 10 Fold is ',acc_cv_log)
print('Ruuning time is ',datetime.timedelta(seconds=GB_time))
```

```python
cat_features=np.where(X_train.dtypes!=np.float)[0]
```

```python
cat_features
```

```python
train_data=Pool(X_train,
                y_train,
                cat_features
)
```

```python
catboost_model = CatBoostClassifier(iterations=1000,
                                    custom_loss=['Accuracy'],
                                    loss_function='Logloss'
                                   )

catboost_model.fit(train_data,plot=True)

acc_catboost=round(catboost_model.score(X_train,y_train)*100,2)
```

```python
acc_catboost
```

```python
start_time=time.time()

cv_params=catboost_model.get_params()

cv_data=cv(train_data,
           cv_params,
           fold_count=10,
           plot=True
)

catboost_time=(time.time()-start_time)

acc_cv_catboost=round(np.max(cv_data['test-Accuracy-mean'])*100,2)
```

```python
print('Accuracy for catboost is: {} '.format(acc_catboost))
print('Accuracy CV 10 Fold for catboost is: {} '.format(acc_cv_catboost))
print('Running time for catboost is: {} '.format(datetime.timedelta(seconds=catboost_time)))
```
